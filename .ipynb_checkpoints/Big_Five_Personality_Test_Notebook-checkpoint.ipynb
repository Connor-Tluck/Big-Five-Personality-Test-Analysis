Openness - 
People who like to learn new things and enjoy new experiences usually score high in openness. Openness includes traits like being insightful and imaginative and having a wide variety of interests.

Conscientiousness - 
People that have a high degree of conscientiousness are reliable and prompt. Traits include being organized, methodic, and thorough.

Extraversion - 
Extraverts get their energy from interacting with others, while introverts get their energy from within themselves. Extraversion includes the traits of energetic, talkative, and assertive.

Agreeableness - 
These individuals are friendly, cooperative, and compassionate. People with low agreeableness may be more distant. Traits include being kind, affectionate, and sympathetic.

Neuroticism - Neuroticism is also sometimes called Emotional Stability. This dimension relates to one’s emotional stability and degree of negative emotions. People that score high on neuroticism often experience emotional instability and negative emotions. Traits include being moody and tense.

```python
#lets train a model to predict the relationship between 
#agreeableness and Extraversion
#https://www.udemy.com/course/python-for-data-science-and-machine-learning-bootcamp/learn/lecture/5733434#overview

tallied_df.columns
```

```python
#lets try to predict agreeableness based 
#off of the other peronality trait results.

#predicted value
y = tallied_df['Agreeableness']

#numerical features
X = tallied_df[['Extraversion', 'Neuroticism', 'Conscientiousness', 'Openness']]
```

```python
#import model_selection
from sklearn.model_selection import train_test_split
```

```python
#create train and test groups ffrom data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#train model on training data
from sklearn.linear_model import LinearRegression
```

```python
#create and instance of a LinearRegression() model and fit to train data.
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
#coefficients will show overall correclation here. As a 1:1 would be a perfect correclation. 
lm.coef_
```

```python
#predict X values by using X_test set. For testing of model. 
predictions = lm.predict(X_test)
```

```python
#model is very inaccurate.
ax = sns.scatterplot(x=y_test,y=predictions,s=20)
ax.axes.set_title("Correlation Heat Map",fontsize=10,pad=20)
ax.set_xlabel("Y Test(True Values)",fontsize=10)
ax.set_ylabel("Predicted Values",fontsize=10)
```

```python
from sklearn import metrics
print('MAE', metrics.mean_absolute_error(y_test,predictions))
print('MSE', metrics.mean_squared_error(y_test,predictions))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
#we can explain about 33 percent of the variance, terrible model. 
metrics.explained_variance_score(y_test,predictions)
```

```python
#the way to interpret this is as follows:
#with all other factors fixed
#one unit increase in agreeableness results in a coeff increase in trait. 

#overall there is a very weak correlation between traits. 

cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```
